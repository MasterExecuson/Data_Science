```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
diretorio_repositorio = 'database'
dataframes = []
```

```python
for root, dirs, files in os.walk(diretorio_repositorio):
    for file in files:
        if file.endswith('.csv'):
            nome_arquivo, extensao = os.path.splitext(file)
            if nome_arquivo[-1] == 'b':
                arquivo_csv = os.path.join(root, file)
                dataframe = pd.read_csv(arquivo_csv, sep=';')
                dataframe = dataframe.transpose()
                dataframe['classe'] = nome_arquivo
                dataframes.append(dataframe)

dataset = pd.concat(dataframes, ignore_index=True)
```
